```python
from sklearn.cluster import KMeans
```

```python
from matplotlib import pyplot as plt
```

```python
import warnings
warnings.filterwarnings("ignore")
import numpy as np

X = np.array([
    [1, 2], [2, 1], [3, 4],
    [5, 7], [6, 8], [7, 6],
    [8, 5], [4, 3]
])

wcss = []

for k in range(1,9):
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(X)
    wcss.append(km.inertia_)

plt.plot(range(1,9), wcss, marker='o')
plt.xlabel("Number of Clusters (K)")
plt.ylabel("WCSS")
plt.title("Elbow Method")
plt.grid(True)
plt.show()
```

```python
import numpy as np
from sklearn_extra.cluster import KMedoids

X = np.array([
    [1, 2], [2, 1], [1, 1], [2, 2],
    [6, 7], [7, 6], [6, 6], [7, 7],
    [3, 8], [4, 9]
])


kmedoids = KMedoids(n_clusters=3, random_state=42)  # metric='manhattan'
kmedoids.fit(X)


print("Medoid indices:", kmedoids.medoid_indices_)
print("Medoids:")
print(kmedoids.cluster_centers_)

print("\nCluster labels:")
print(kmedoids.labels_)

plt.scatter(X[:, 0], X[:, 1], label='Data points')
plt.scatter(kmedoids.cluster_centers_[:, 0],
            kmedoids.cluster_centers_[:, 1], color='green')

plt.show()
```

```python
import numpy as np
from sklearn_extra.cluster import KMedoids

X = np.array([[8, 4],
 [4, 2],
 [3, 4],
 [3, 5],
 [1, 4],
 [1, 8],
 [3, 7],
 [5, 2],
 [4, 7],
 [9, 5],
 [6, 2],
 [1, 4],
 [2, 2],
 [8, 8],
 [6, 2]]
)


kmedoids = KMedoids(n_clusters=3, random_state=42) 


kmedoids.fit(X)

print("Medoid indices:", kmedoids.medoid_indices_)
print("Medoids:")
print(kmedoids.cluster_centers_)

print("\nCluster labels:")
print(kmedoids.labels_)

plt.scatter(X[:, 0], X[:, 1], label='Data points')
plt.scatter(kmedoids.cluster_centers_[:, 0], kmedoids.cluster_centers_[:, 1],marker='x', color='yellow')
plt.grid()
plt.show()

```

```python
kmeans=KMeans(n_clusters=3,random_state=42)
labels=kmeans.fit_predict(X)
plt.scatter(X[:,0],X[:,1],c=labels)
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],marker='x',color='red')
print("centroids: ",kmeans.cluster_centers_)
plt.grid()
plt.show()
```
