```python
import numpy as np
X = np.array([[1,2], [2,3], [3,4], [4,5]])
Y = np.array([0, 0, 1, 1])
centers = np.array([[1,2], [3,4]])
sigma = 1.0

def rbf(x,c):
    return np.exp(-np.linalg.norm(x - c)**2 /(2 * sigma**2))

H = np.array([[rbf(x, c) for c in centers] for x in X])

w = np.linalg.pinv(H) @ Y

y_pred = H @ w
y_pred_class = (y_pred > 0.5).astype(int)

print("Predicted: ", y_pred_class)
```

```python
import numpy as np
import matplotlib.pyplot as plt
X = np.array([[1], [2], [3], [4]])
Y = np.array([0, 0, 1, 1])
centers = np.array([[1], [4]])
sigma = 1.0

def rbf(x,c):
    return np.exp(-np.linalg.norm(x - c)**2 /(2 * sigma**2))

H = np.array([[rbf(x, c) for c in centers] for x in X])

w = np.linalg.pinv(H) @ Y

y_pred = H @ w
y_pred_class = (y_pred > 0.5).astype(int)

print("Predicted: ", y_pred)
print("Predicted Class: ", y_pred_class)
x_vals = np.linspace(0, 5, 100)

H = np.array([[rbf(x, c) for c in centers] for x in x_vals])

plt.figure()
plt.plot(x_vals, H[:, 0], label ="Center = 1")
plt.plot(x_vals, H[:, 1], label ="Center = 4")

plt.scatter(X, [0]*len(X))

plt.title("RBF Curves using H matrix")
plt.xlabel("Input (x)")
plt.ylabel("RBF Output")
plt.legend()
plt.grid(True)

plt.show()
```

```python
# Simple transactions
transactions = [
    {"A", "B"},
    {"A", "C"},
    {"B", "C"},
    {"A", "B"}
]

min_sup = 2

# Step 1: Build TID Lists (vertical format)
tid = {}

for i, t in enumerate(transactions):
    for item in t:
        if item not in tid:
            tid[item] = set()
        tid[item].add(i)

print("TID Lists:", tid)

# Step 2: Intersect manually (example: A and B)
AB = tid["A"] & tid["B"]
BC = tid["B"] & tid["C"]
AC = tid["A"] & tid["C"]# core idea of ECLAT
print("Support of (A,B):", len(AB))
print("Support of (B,C):", len(BC))
print("Support of (A,C):", len(AC))

# Step 3: Check if frequent
if len(AB) >= min_sup:
    print("(A,B) is frequent")
else:
    print("(A,B) is frequent")
if len(BC) >= min_sup:
    print("(B,C) is not frequent")
else:
    print("(B,C) is not frequent")
if len(AC) >= min_sup:
    print("(A,C) is not frequent")
else:
    print("(A,C) is not frequent")
```
