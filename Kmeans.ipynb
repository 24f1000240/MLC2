```python
import numpy as np
import pandas as pd
```

```python
data=np.random.rand(10, 2)*10
data
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(data[:, 0], data[:, 1])
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Scatter plot befre clustering")
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
labels=kmeans.fit_predict(data)
```

```python
plt.scatter(data[:, 0], data[:, 1],c=labels)
plt.scatter(kmeans.cluster_centers_[:,0],
           kmeans.cluster_centers_[:,1],
           marker='x')
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Scatter plot after clustering")
plt.grid(True)
plt.show()
```

```python
print("Centroid are \n",kmeans.cluster_centers_)
```

```python
import pandas as pd
df=pd.read_csv("Dataset_spine.csv")
point=df[['Col1','Col2']].values
kmeans=KMeans(n_clusters=5, random_state=42)
labels=kmeans.fit_predict(point)
#plt.scatter(x,y,c=labels)
plt.scatter(point[:,0],point[:,1],c=labels)
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],marker='x',color='red')
print("centroids:",kmeans.cluster_centers_)
plt.grid()
plt.show()
```
