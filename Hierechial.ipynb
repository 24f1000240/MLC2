```python
import warnings
warnings.filterwarnings("ignore")
import numpy as np
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

X = np.array([
    [1, 2], [2, 1], [1, 1], [2, 2],
    [6, 7], [7, 6], [6, 6], [7, 7]
])

Z = linkage(X, method='complete')  # computes hierarchical clustering result
print(Z)

dendrogram(Z)  # visualizes the hierarchical structure; leaves = data points
plt.title("Hierarchical Clustering Dendrogram")
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.show()
```

```python
import warnings
warnings.filterwarnings("ignore")
import numpy as np
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from scipy.cluster.hierarchy import fcluster

X = np.array([
    [1, 2], [2, 1], [1, 1], [2, 2],
    [6, 7], [7, 6], [6, 6], [7, 7]
])

Z = linkage(X, method='complete')  # computes hierarchical clustering result
print(Z)
cluster_labels=fcluster(Z,t=4,criterion="maxclust")
print(cluster_labels)
plt.scatter(X[:,0],X[:,1],c=cluster_labels)
plt.title("Hierarchical Clustering FCLUSTER")
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.show()
```

```python
import warnings
warnings.filterwarnings("ignore")
import numpy as np
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from scipy.cluster.hierarchy import fcluster

X = np.array([
    [1, 2], [2, 1], [1, 1], [2, 2],
    [6, 7], [7, 6], [6, 6], [7, 7]
])

Z = linkage(X, method='single')  # computes hierarchical clustering result
print(Z)
cluster_labels=fcluster(Z,t=1.2,criterion="distance")
print(cluster_labels)
plt.scatter(X[:,0],X[:,1],c=cluster_labels)
plt.title("Hierarchical Clustering FCLUSTER")
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.show()
```
