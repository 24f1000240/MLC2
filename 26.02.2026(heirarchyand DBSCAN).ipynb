```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

X = np.array([
    [1, 2], [2, 1], [1, 1], [2, 2],
    [6, 7], [7, 6], [6, 6], [7, 7]
])

Z = linkage(X, method='complete')  # computes hierarchical clustering result
print(Z)

dendrogram(Z)  # visualizes the hierarchical structure; leaves = data points
#By axis= cluster distance
plt.title("Hierarchical Clustering Dendrogram")
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.show()
X.shape
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

X = np.array([
    [1, 2], [2, 1], [1, 1], [2, 2],
    [6, 7], [7, 6], [6, 6], [7, 7]
])

Z = linkage(X, method='ward')  # computes hierarchical clustering result
print(Z)

dendrogram(Z)  # visualizes the hierarchical structure; leaves = data points
#By axis= cluster distance
plt.title("Hierarchical Clustering Dendrogram")
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.show()
X.shape
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import fcluster

X = np.array([
    [1, 2], [2, 1], [1, 1], [2, 2],
    [6, 7], [7, 6], [6, 6], [7, 7]
])

Z = linkage(X, method='complete')  # computes hierarchical clustering result
print(Z)


cluster_labels=fcluster(Z,t=4,criterion='maxclust')
print(cluster_labels)
plt.scatter(X[:,0],X[:,1],c=cluster_labels)
plt.title("Hierarchical Clustering FCluster")
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.show()
X.shape
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import fcluster

X = np.array([
    [1, 2], [2, 1], [1, 1], [2, 2],
    [6, 7], [7, 6], [6, 6], [7, 7]
])

Z = linkage(X, method='complete')  # computes hierarchical clustering result
print(Z)


cluster_labels=fcluster(Z,t=1.2,criterion='distance')
print(cluster_labels)
plt.scatter(X[:,0],X[:,1],c=cluster_labels)
plt.title("Hierarchical Clustering FCluster")
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.show()
X.shape
```

```python
# DENSITY-BASED CLUSTERING (DBSCAN)
# DBSCAN imported only when needed
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps=1.5, min_samples=2)

# eps = radius to consider neighbors.
# min_samples = minimum points to form a cluster.

db_labels = dbscan.fit_predict(X)
print(db_labels)

# Points labeled -1 are considered noise/outliers.
plt.scatter(X[:, 0], X[:, 1], c=db_labels)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("DBSCAN Clustering")

plt.show()
```
